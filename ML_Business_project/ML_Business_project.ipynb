## Training and Testing the model for each region:

```python
# Function to train and evaluate the model
def train_and_evaluate(region_data, region_name):
    # Drop 'id' column and split into features and target
    features = region_data.drop(columns=['id', 'product'])
    target = region_data['product']

    # Split data into training (75%) and validation (25%)
    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.25, random_state=42
    )

    # Train Linear Regression model
    model = LinearRegression()
    model.fit(features_train, target_train)

    # Make predictions
    predictions = model.predict(features_valid)

    # Calculate RMSE
    rmse = np.sqrt(mean_squared_error(target_valid, predictions))

    # Print results
    print(f"\nRegion {region_name}:")
    print(f"Average predicted reserves: {predictions.mean():.2f} thousand barrels")
    print(f"RMSE: {rmse:.2f} thousand barrels")

    return predictions, target_valid

# Train and evaluate model for each region
predictions_0, target_valid_0 = train_and_evaluate(geo_data_0, "0")
predictions_1, target_valid_1 = train_and_evaluate(geo_data_1, "1")
predictions_2, target_valid_2 = train_and_evaluate(geo_data_2, "2")
```

The results from the model evaluation indicate significant differences in prediction accuracy across the three regions. **Region 1** demonstrates the lowest error, with an **RMSE of 0.89 thousand barrels**, suggesting that the model performs exceptionally well in this region and predictions are highly reliable. However, **Region 0 and Region 2** have considerably higher errors, with **RMSE values of 37.76 and 40.15 thousand barrels, respectively**, indicating a greater degree of variability and uncertainty in the predictions. Despite these differences, the average predicted reserves are **92.40 thousand barrels for Region 0, 68.71 thousand barrels for Region 1, and 94.77 thousand barrels for Region 2**. While **Regions 0 and 2** have higher predicted reserves, their high RMSE values introduce greater uncertainty, making their profitability less predictable. **Region 1, although having the lowest predicted reserves, provides the most stable and accurate predictions**, which could be a crucial factor when assessing financial risks and selecting the optimal drilling location.

## Selecting the Best Wells

```python
def calculate_profit(predictions, target, region_name, verbose=True):
    """
    Function to calculate profit for the top 200 predicted wells.
    """
    # Convert to Pandas Series with unique indices
    predictions = pd.Series(predictions).reset_index(drop=True)
    target = pd.Series(target).reset_index(drop=True)

    # Select top 200 wells based on predicted reserves
    top_wells = predictions.sort_values(ascending=False).head(200)

    # Calculate total actual reserves for selected wells
    total_actual_reserves = target[top_wells.index].sum()

    # Calculate profit
    total_revenue = total_actual_reserves * BARREL_REVENUE  # Total revenue from selected wells
    profit = total_revenue - BUDGET  # Deduct the development budget

    if verbose:  # Print only when verbose=True
        print(f"\nRegion {region_name}:")
        print(f"Total predicted reserves for top 200 wells: {top_wells.sum():.2f} thousand barrels")
        print(f"Total actual reserves from selected wells: {total_actual_reserves:.2f} thousand barrels")
        print(f"Expected profit: ${profit / 1e6:.2f} million")

    return profit
# Calculate profit for each region
profit_0 = calculate_profit(predictions_0, target_valid_0, "0")
profit_1 = calculate_profit(predictions_1, target_valid_1, "1")
profit_2 = calculate_profit(predictions_2, target_valid_2, "2")
```

## Implementing Bootstrapping

```python
# Bootstrapping function using `calculate_profit`
def bootstrap_profit_analysis(predictions, target, region_name, n_samples=1000):
    """
    Perform bootstrapping to analyze profit distribution using the calculate_profit function.
    """
    # Ensure predictions and target are Pandas Series with unique indices
    predictions = pd.Series(predictions).reset_index(drop=True)
    target = pd.Series(target).reset_index(drop=True)

    profits = []

    for _ in range(n_samples):
        # Sample 500 wells by position with replacement
        sampled_indices = randint(0, len(target), size=500)

        # Use `calculate_profit` with `verbose=False` to avoid excessive printing
        profit = calculate_profit(predictions.iloc[sampled_indices], target.iloc[sampled_indices], region_name=None, verbose=False)
        profits.append(profit)

    # Convert profits to NumPy array
    profits = np.array(profits)

    # Calculate statistics
    mean_profit = profits.mean()
    lower_bound = np.percentile(profits, 2.5)  # 2.5% percentile
    upper_bound = np.percentile(profits, 97.5)  # 97.5% percentile
    risk_of_loss = (profits < 0).mean() * 100  # Percentage of negative profit

    # Print final bootstrapping results
    print(f"\nRegion {region_name} - Bootstrapping Analysis:")
    print(f"Average Expected Profit: ${mean_profit / 1e6:.2f} million")
    print(f"95% Confidence Interval: (${lower_bound / 1e6:.2f}M, ${upper_bound / 1e6:.2f}M)")
    print(f"Risk of Loss: {risk_of_loss:.2f}%")

    return mean_profit, lower_bound, upper_bound, risk_of_loss

# Run bootstrapping for each region
mean_profit_0, lower_0, upper_0, risk_0 = bootstrap_profit_analysis(predictions_0, target_valid_0, "0")
mean_profit_1, lower_1, upper_1, risk_1 = bootstrap_profit_analysis(predictions_1, target_valid_1, "1")
mean_profit_2, lower_2, upper_2, risk_2 = bootstrap_profit_analysis(predictions_2, target_valid_2, "2")
```
