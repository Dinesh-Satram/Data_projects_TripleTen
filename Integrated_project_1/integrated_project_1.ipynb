## Analysis of Data

### Game Releases over time

```python
# Count the number of games released each year
games_per_year = data['year_of_release'].value_counts().sort_index()

# Plot the number of games released each year
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
games_per_year.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Games Released Per Year', fontsize=14)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Number of Games', fontsize=12)
plt.xticks(rotation=45)  # Rotate the x-axis labels by 45 degrees
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The data shows low game releases in the 1980s and early 1990s, with a significant rise from 2000 to 2009. The peak occurs around 2007-2009, after which releases stabilize in the 2010s but remain relatively high. The earlier periods are less significant in terms of release volume.

### Platform sales analysis

```python
# Calculate total sales per platform
platform_sales = data.groupby('platform')['total_sales'].sum().sort_values(ascending=False)

# Plot total sales by platform
platform_sales.plot(kind='bar', figsize=(12, 6), color='orange', edgecolor='black')
plt.title('Total Sales by Platform', fontsize=14)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Total Sales (millions)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Identify the top platforms with the highest sales
top_platforms = platform_sales.head(5).index

# Sales trends for top platforms over time
for platform in top_platforms:
    yearly_sales = data[data['platform'] == platform].groupby('year_of_release')['total_sales'].sum()
    plt.plot(yearly_sales, label=platform)

plt.title('Sales Trends of Top Platforms Over Time', fontsize=14)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Total Sales (millions)', fontsize=12)
plt.legend(title='Platform')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The platforms with the highest total sales, like "PS2", "X360", and "Wii", saw rapid growth and peaked around 2005-2010. After reaching their heights, they experienced a sharp decline, with "PS2" fading around 2010. Older platforms like "PSV" and "3DO" now have zero sales, showing they became obsolete. New platforms generally take around 5-7 years to peak in sales, after which they decline as newer generations take over. This cycle is evident in the graphs, where the rise and fall of platforms follow a similar pattern.

## Test Hypothesis

```python
from scipy import stats

# Hypothesis 1: Test if average user ratings for Xbox One and PC are the same

# Null Hypothesis (Hâ‚€): Average user ratings for Xbox One and PC are the same.
# Alternative Hypothesis (Hâ‚): Average user ratings for Xbox One and PC are different.

# Extract user ratings for Xbox One and PC
xbox_one_ratings = relevant_data[relevant_data['platform'] == 'XOne']['user_score'].dropna()
pc_ratings = relevant_data[relevant_data['platform'] == 'PC']['user_score'].dropna()

# Perform an independent two-sample t-test
t_stat_1, p_value_1 = stats.ttest_ind(xbox_one_ratings, pc_ratings)

# Output results
print(f"T-statistic for Xbox One vs PC: {t_stat_1}")
print(f"P-value for Xbox One vs PC: {p_value_1}")

# Hypothesis 2: Test if average user ratings for Action and Sports genres are different

# Null Hypothesis (Hâ‚€): Average user ratings for Action and Sports genres are the same.
# Alternative Hypothesis (Hâ‚): Average user ratings for Action and Sports genres are different.

# Extract user ratings for Action and Sports genres
action_ratings = relevant_data[relevant_data['genre'] == 'Action']['user_score'].dropna()
sports_ratings = relevant_data[relevant_data['genre'] == 'Sports']['user_score'].dropna()

# Perform an independent two-sample t-test
t_stat_2, p_value_2 = stats.ttest_ind(action_ratings, sports_ratings)

# Output results
print(f"T-statistic for Action vs Sports: {t_stat_2}")
print(f"P-value for Action vs Sports: {p_value_2}")

# Chosen significance level (alpha) = 0.05
alpha = 0.05

# Conclusion for hypothesis 1
if p_value_1 < alpha:
    print("Reject the null hypothesis: There is a significant difference between Xbox One and PC ratings.")
else:
    print("Fail to reject the null hypothesis: No significant difference between Xbox One and PC ratings.")

# Conclusion for hypothesis 2
if p_value_2 < alpha:
    print("Reject the null hypothesis: The data provides sufficient evidence to conclude there is a significant difference between Action and Sports ratings.")
else:
    print("Fail to reject the null hypothesis: No significant difference between Action and Sports ratings.")
```

The hypothesis test for average user ratings between Xbox One and PC yielded a **T-statistic** of 0.4575 and a **P-value** of 0.6476, which is greater than the significance level of 0.05. Therefore, we **fail to reject** the null hypothesis, indicating that there is **no significant difference** in the average user ratings between Xbox One and PC platforms. On the other hand, the hypothesis test comparing user ratings for the Action and Sports genres returned a **T-statistic** of 7.8917 and a **P-value** of 1.4973e-14, which is far below the significance level. As a result, we **reject the null hypothesis** and conclude that there is a **significant difference** between the average user ratings for Action and Sports genres. This indicates that user perception differs significantly between these two genres.
